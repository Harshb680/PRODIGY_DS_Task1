```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("F:/Prodigy/IPynb files/TrainTitanic.csv",index_col=0)
df
```

```python
#Cleaning Data
df.isnull().sum()
```

```python
#Filling NaN values of Age with Mean 
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
#Filling NaN values of Embarked with Mode 
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
df
```

```python
#Droping Cabin Column as more than 70% data is NaN values 
df.drop(columns=["Cabin"],inplace=True)
df
```

```python
df.isna().sum()
```

```python
#Basic Exploration on the dataset
df.info()   # Check data types and missing values
df.describe()  # Summary statistics of numerical columns
df.nunique()  # Check unique values in each column
```

```python
#Basic EDA
df["Survived"].value_counts(normalize=True)*100
```

```python
#Survived Vs Sex
sns.barplot(x="Sex",y="Survived",data=df)
plt.show()
```

```python
#Age Distribution
sns.histplot(df['Age'], bins=30, kde=True)  
plt.title("Age Distribution of Passengers")  
plt.show()
```

```python
#Relationship between Survival & Gender
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Relationship between Survival & Gender")
plt.show()
```

```python
#Survival Rate by Passenger Class (Pclass)
sns.barplot(x="Pclass", y="Survived", data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
#To check if traveling alone was a disadvantage:
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1  # +1 to include the passenger  
sns.barplot(x="FamilySize", y="Survived", data=df)
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.title("Effect of Family Size on Survival")
plt.show()
```

```python
#To analyze if richer passengers had better survival rates
plt.figure(figsize=(8,5))
sns.boxplot(x="Survived", y="Fare", data=df)
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.title("Fare Paid vs. Survival")
plt.show()
```
