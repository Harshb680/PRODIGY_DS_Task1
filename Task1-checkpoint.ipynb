```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample explicit/non-explicit counts
explicit_counts = {'Non-Explicit': 105000, 'Explicit': 12000}  # Replace with actual counts

# Convert data into a DataFrame
df_plot = pd.DataFrame({
    'Explicit Content': list(explicit_counts.keys()),
    'Count': list(explicit_counts.values())
})

# Define colors manually
colors = ['#4C72B0', '#DD8452']  # Blue and Orange

# Set style
sns.set_style("darkgrid")

# Plot
plt.figure(figsize=(7, 5))
ax = sns.barplot(data=df_plot, x='Explicit Content', y='Count', hue='Explicit Content', 
                 palette=colors, legend=False, width=0.5)  # Reduced width for better spacing

# Adding data labels inside the bars
for i, count in enumerate(df_plot['Count']):
    ax.text(i, count + 3000, f"{count:,}", ha='center', fontsize=12, color='black', fontweight='bold')

# Adjust y-axis to prevent labels from going outside
plt.ylim(0, max(explicit_counts.values()) * 1.3)

# Labels and Title
plt.xlabel("Explicit Content", fontsize=14, fontweight='bold')
plt.ylabel("Number of Songs", fontsize=14, fontweight='bold')
plt.title("Distribution of Explicit vs. Non-Explicit Songs", fontsize=16, fontweight='bold')

# Grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
